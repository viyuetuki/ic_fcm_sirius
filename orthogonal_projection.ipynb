# Orthogonal Projection

Geometric view

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Plotting a vector

# arrow point from (x,y) to (x+u, y+v)

# origin = (0,0)
# v1 = (2,3)

X1 = [0] #first x coordinate ("origin")
Y1 = [0] #first y coordinate ("origin")
U1 = [2] #second x coordinate ("direction")
V1 = [3] #second y coordinate ("direction")
plt.quiver(X1, Y1, U1, V1, 
           color='b', 
           angles='xy', scale_units='xy', scale=1)
plt.scatter(2, 3, color='b',label='v1')
#plt.scatter(X1, Y1, s=100, color='b') 
#plt.title('Single Vector') 
plt.xlim(0, 5) 
plt.ylim(0, 5) 
plt.grid('on') 
plt.legend()
plt.show() 
```

```python
#Plotting two vectors

# arrows point from (x,y) to (x+u, y+v)

# origins = (0,0)
# v1 = (2,3)
# v2 = (4,0.8)

X = [0,0] #coordinate x of the origin of each vector
Y = [0,0] #coordinate y of the origin of each vector
U = [2,4] #coordinate x of the direction of each vector  
V = [3,0.8] #coordinate y of the direction of each vector  
plt.quiver(X, Y, U, V, 
           color=['b', 'm'], 
           angles='xy', scale_units='xy', scale=1)
plt.scatter(2, 3, color='b',label='v1')
plt.scatter(4, 0.8, color='m',label='v2')
#plt.scatter(X, Y, s=100, color=['b', 'm']) 
#plt.title('Single Vector') 
plt.xlim(0, 5) 
plt.ylim(0, 5) 
plt.grid('on')
plt.legend() 
plt.show() 
```

Let's consider any vectors $\vec{v_{1}}$ and $\vec{v_{2}}$, and the angle $\theta$ between them.

We can call $\hat{\vec{v_{1}}}$ the orthogonal projection of the vector $\vec{v_{1}}$ onto $\vec{v_{2}}$. So, $\hat{\vec{v_{1}}}$ is a parallel vector to $\vec{v_{2}}$, but is not the same. It can be defined as: $$\hat{\vec{v_{1}}} = \alpha \vec{v_{2}}$$

The orthogonal vector to $\vec{v_{2}}$ can be determined as $\vec{z}$. 

Then, we can also say that: $$\vec{v_{1}} = \hat{\vec{v_{1}}} + \vec{z}$$ $$ \hat{\vec{v_{1}}} = ||v_{1}||cos(\theta) $$ $$ \vec{z} = ||v_{1}|| sen(\theta)$$


Taking everything into account, to project $\vec{v_{1}}$ orthogonally to $\vec{v_{2}}$, our main goal is to find out the $\alpha$ factor.

Remembering that $\vec{z}$ is orthogonal to $\vec{v_{2}}$, we have that: $$ \vec{z} \cdot \vec{v_{2}} = 0$$ $$ (\vec{v_{1}} - \hat{\vec{v_{1}}}) \cdot \vec{v_{2}} = 0 $$ $$  (\vec{v_{1}} - \alpha \vec{v_{2}}) \cdot \vec{v_{2}} = 0 $$ $$ \vec{v_{1}} \cdot \vec{v_{2}} = \alpha \vec{v_{2}} \cdot \vec{v_{2}} $$

Isolating $\alpha$: $$ \alpha = \frac{\vec{v_{1}} \cdot \vec{v_{2}}}{\vec{v_{2}} \cdot \vec{v_{2}}} $$

Rewriting $\vec{z}$: $$ \vec{z} = \vec{v_{1}} - \big( \frac{\vec{v_{1}} \cdot \vec{v_{2}}}{\vec{v_{2}} \cdot \vec{v_{2}}} \big) \vec{v_{2}} $$

```python
#Calculating "alpha"
v1 = np.array([2, 3])
v2 = np.array([4, 0.8])


alpha = (np.inner(v1,v2))/(np.inner(v2,v2))
print("α =", alpha)
```

```python
#Finding the vector z
z = v1 - alpha*v2
print("z =",z)
```

```python
X = [0,0, 0] 
Y = [0,0, 0] 
U = [2,4, -0.5]   
V = [3,0.8, 2.5]   
plt.quiver(X, Y, U, V, 
           color=['b', 'm', 'r'], 
           angles='xy', scale_units='xy', scale=1)
#plt.scatter(X, Y, s=100, color=['b', 'm']) 
#plt.title('Single Vector') 
plt.scatter(2, 3, color='b',label='v1')
plt.scatter(4, 0.8, color='m',label='v2')
plt.scatter(-0.5,2.5, color='r',label='z')
plt.xlim(-2, 5) 
plt.ylim(-2, 5) 
plt.grid('on') 
plt.legend()
plt.show() 
```

```python
#Finally, finding out the orthogonal projection of v1 into v2, 
# v1_hat
v1_hat = alpha*v2
print("v1_hat =", v1_hat)
```

```python
#Plotting all together
X = [0,0,2.5,0] 
Y = [0,0,0.5,0] 
U = [2,4, -0.5,2.5]   
V = [3,0.8, 2.5,0.5]   
plt.quiver(X, Y, U, V, 
           color=['b', 'm', 'r','purple'], 
           angles='xy', scale_units='xy', scale=1)
#plt.scatter(X, Y, s=100, color=['b', 'm']) 
#plt.title('Single Vector') 
plt.scatter(2, 3, color='b',label='v1')
plt.scatter(4, 0.8, color='m',label='v2')
plt.scatter(2,3, color='r',label='z')
plt.scatter(2.5, 0.5, color='purple',label='v1_hat')
plt.xlim(-2, 5) 
plt.ylim(-2, 5) 
plt.grid('on') 
plt.legend()
plt.show() 
```
